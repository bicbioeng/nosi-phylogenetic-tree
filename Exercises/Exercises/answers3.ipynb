```python
from Bio import Phylo
from io import StringIO
import matplotlib.pyplot as plt

# Newick format for a cladogram (branching order only)
cladogram = "(((Human, Chimpanzee), Gorilla), Orangutan);"
tree = Phylo.read(StringIO(cladogram), "newick")

# Visualization
fig, ax = plt.subplots(figsize=(10, 8))
Phylo.draw(tree, axes=ax)
ax.set_title("Cladogram: Vertebrate Evolution", fontsize=14, weight='bold')
plt.show()
```

### Cladogram Structure
The structure of the tree can be represented as:

- Human and Chimpanzee are closely related and form a clade (i.e., they share a common ancestor more recently).
- This clade (Human + Chimpanzee) is next most closely related to Gorilla, forming another clade.
- The larger clade containing (Human, Chimpanzee, and Gorilla) is next most closely related to Orangutan.

The most closely related species in the cladogram are Human and Chimpanzee, as they share the most recent common ancestor within this structure.
